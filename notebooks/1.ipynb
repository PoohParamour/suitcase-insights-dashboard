```python
# STEP 1 : Setup & Data Load
import pandas as pd
import numpy as np

# 1.1 โหลดไฟล์
FILE_PATH = "../data/data_renamefinal.csv"      # เปลี่ยน path ได้ถ้าไฟล์อยู่อีกที่
df = pd.read_csv(FILE_PATH)

# 1.2 ดูรูปทรงข้อมูลเบื้องต้น
print("Shape:", df.shape)               # จำนวนแถว × คอลัมน์
display(df.head())                      # 5 แถวแรก
display(df.tail(3))                     # 3 แถวท้าย

# 1.3 สรุปชนิดข้อมูล + ช่องโหว่
display(df.info())                      # dtype + non-null
missing = df.isna().sum().sort_values(ascending=False)
display(missing[missing > 0].head(20))  # คอลัมน์ที่มี Missing มากที่สุด

from io import StringIO
buf = StringIO()
df.info(buf=buf)
with open("01_data_overview.txt", "w", encoding="utf-8") as f:
    f.write(buf.getvalue())


```

```python
# STEP 2 : Cleaning & Type-casting
import re
from io import StringIO

# 2.1 กำจัดแถวที่ demographic สำคัญหาย (เพศ อายุ รายได้)
df = df.dropna(subset=["gender", "age", "monthly_income"]).reset_index(drop=True)

# 2.2 แปลง age -> int
df["age"] = df["age"].astype(int)

# 2.3 แปลงช่วงรายได้เป็น midpoint (ตัวเลข)
def income_to_midpoint(x):
    # ตัวอย่างรูป "15,001 - 20,000"
    nums = list(map(int, re.findall(r"\d+", x.replace(",", ""))))
    return np.mean(nums) if nums else np.nan

df["income_mid"] = df["monthly_income"].apply(income_to_midpoint)

# 2.4 เติมค่า missing ให้สเกล 1-5 ด้วย median ของคอลัมน์
likert_cols = [c for c in df.columns if re.match(r"(factor|price|channel|promo)_", c)]
for col in likert_cols:
    df[col] = df[col].fillna(df[col].median())

# 2.5 จัดประเภทอายุ & รายได้ (ไว้ทำแผนภูมิ)
df["age_group"] = pd.cut(
    df["age"], bins=[0,17,24,34,44,150],
    labels=["<18","18-24","25-34","35-44","45+"]
)
df["income_group"] = pd.cut(
    df["income_mid"], bins=[0,15000,30000,50000,1e9],
    labels=["<15k","15-30k","30-50k","50k+"]
)

# 2.6 เซฟชุดข้อมูลสะอาดเผื่อใช้ต่อ
CLEAN_PATH = "clean_survey.parquet"
df.to_parquet(CLEAN_PATH, index=False)
print(f"✅ Saved cleaned data to {CLEAN_PATH}   |  Shape: {df.shape}")
```

```python
# STEP 3 : Demographic × Importance Factors
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# 3.1 กำหนดคอลัมน์สเกล 1-5
factor_cols = [c for c in df.columns if c.startswith("factor_")]

# ---------- 3A. Pivot Table ตามเพศ ----------
gender_factor = (
    df.groupby("gender")[factor_cols]
      .mean()
      .round(2)
      .sort_values("factor_durability", ascending=False)
)
display(gender_factor.style.background_gradient(cmap="Greens", axis=1))

# ---------- 3B. Heat-map ตามช่วงอายุ ----------
age_factor = df.groupby("age_group")[factor_cols].mean()
plt.figure(figsize=(10, 6))
sns.heatmap(age_factor, annot=True, cmap="YlGnBu", vmin=1, vmax=5)
plt.title("Average Importance Score by Age Group")
plt.ylabel("Age Group")
plt.xlabel("Importance Factors (1–5)")
plt.tight_layout()
plt.show()

# ---------- 3C. Heat-map ตามช่วงรายได้ ----------
income_factor = df.groupby("income_group")[factor_cols].mean()
plt.figure(figsize=(10, 6))
sns.heatmap(income_factor, annot=True, cmap="OrRd", vmin=1, vmax=5)
plt.title("Average Importance Score by Income Group")
plt.ylabel("Income Group")
plt.xlabel("Importance Factors (1–5)")
plt.tight_layout()
plt.show()

# ---------- 3D. ANOVA Example ----------
pval = stats.f_oneway(
    df.loc[df.gender=="ชาย", "factor_weight"],
    df.loc[df.gender=="หญิง", "factor_weight"]
).pvalue
print(f"ANOVA p-value : factor_weight vs gender  →  {pval:.4f}")
```

```python
# ---------- Style / Color (clean & robust) ----------
import unicodedata, re
def normalize_th(s):
    """ลบช่องว่างซ้ำ + แปลง full-width → half-width"""
    if pd.isna(s):
        return s
    s = unicodedata.normalize("NFKC", s)          # normalize unicode
    s = re.sub(r"\s+", " ", s)                    # collapse whitespace
    s = s.strip(" ,;/")                           # ตัด , ; / หน้า-ท้าย
    return s.lower()                              # ทำเป็น lower-case

style_long = (
    df["preferred_styles"]
      .fillna("")
      .apply(normalize_th)
      .str.replace(r"[\[\]\']", "", regex=True)
      .str.replace(r"[;,/|]", ",", regex=True)     # ทุกตัวคั่น → คอมมา
      .str.split(",")
      .explode()
      .str.strip()
      .replace("", pd.NA)
      .dropna()
      .to_frame(name="style")
)

# รวมคำสะกดใกล้เคียง (mapping ย่อ – เติมได้ภายหลัง)
replace_map = {
    "สีเรียบ เท่ (ดำ เทา น้ำเงิน)": "เรียบ เท่",
    "สีพาสเทล น่ารัก": "พาสเทล น่ารัก",
    "สีนูทรา (ทอง เงิน)": "นูทรา (ทอง/เงิน)",
}
style_long["style"] = style_long["style"].replace(replace_map)

style_counts = (
    style_long["style"]
      .value_counts()
      .head(15)
      .rename_axis("Style / Color")
      .reset_index(name="Count")
)

display(style_counts)

plt.figure(figsize=(8,5))
sns.barplot(
    y="Style / Color", x="Count",
    data=style_counts,
    order=style_counts["Style / Color"]
)
plt.title("Top Preferred Styles / Colors")
plt.xlabel("Number of Mentions")
plt.ylabel("")
plt.tight_layout()
plt.show()

print("\n✅  แสดงผล Style/Color สำเร็จ (ถ้าเห็นกราฟและตารางด้านบน)")
```

```python
# === STEP 4B : Product Preferences (full, robust) =========================
import pandas as pd, numpy as np, re, unicodedata
import matplotlib.pyplot as plt, seaborn as sns

##------------------------------------------------------------
## 1) UTILITIES
##------------------------------------------------------------
def explode_multiselect(series, *, sep_pattern=r"[;,/|]", clean_func=None):
    """
    รับ pd.Series (object) ที่อาจมีค่าหลายตัวใน 1 cell
    - sep_pattern  : regex จับตัวคั่นทุกชนิด , ; / |
    - clean_func   : ฟังก์ชันทำความสะอาดคำรายตัว (optional)
    คืน DataFrame 1 คอลัมน์ชื่อ 'value'
    """
    exploded = (
        series.fillna("")
              .astype(str)
              .str.replace(r"[\[\]\']", "", regex=True)      # ตัด [] '
              .str.replace(sep_pattern, ",", regex=True)     # ทุกตัวคั่น → ,
              .str.split(",")
              .explode()
              .str.strip()
              .replace("", pd.NA)
              .dropna()
    )
    if clean_func:
        exploded = exploded.apply(clean_func)
    return exploded.to_frame(name="value")

def normalize_th(text):
    """ลบช่องว่างซ้ำ + normalize unicode + lower-case"""
    text = unicodedata.normalize("NFKC", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text.lower()

def plot_bar(df_counts, *, x="Count", y, title, h=5, w=8, palette="viridis"):
    plt.figure(figsize=(w, h))
    sns.barplot(data=df_counts, x=x, y=y, palette=palette,
                order=df_counts[y])          # เรียงตาม index ปัจจุบัน
    plt.title(title)
    plt.xlabel("Number of Mentions")
    plt.ylabel("")
    plt.tight_layout()
    plt.show()

##------------------------------------------------------------
## 2) BAG TYPES
##------------------------------------------------------------
bag_col = "used_bag_types"
bag_long = explode_multiselect(df[bag_col], clean_func=normalize_th)
top_bags = (bag_long["value"]
            .value_counts()
            .head(10)
            .rename_axis("Bag Type")
            .reset_index(name="Count"))
display(top_bags)
plot_bar(top_bags, y="Bag Type", title="Top Bag Types Used")

# ---------- LUGGAGE SIZE (short trip)  — FIX ----------
size_col = "luggage_size_short_trip"

size_counts = (
    df[size_col]
      .dropna()
      .apply(normalize_th)
      .value_counts()                 # Series
      .rename_axis("Size (inches)")   # ชื่อ index ใหม่
      .reset_index(name="Count")      # ชื่อคอลัมน์ count → Count
)

display(size_counts)

plot_bar(
    size_counts.sort_values("Count", ascending=False),
    y="Size (inches)",
    title="Preferred Size for Short Trips",
    h=4, w=6
)

# ----- แก้บรรทัดดึง Insight ให้ใช้ size_counts.iloc[0]['Size (inches)'] ได้ -----


##------------------------------------------------------------
## 4) STYLE / COLOR
##------------------------------------------------------------
style_col = "preferred_styles"
style_long = explode_multiselect(df[style_col], clean_func=normalize_th)
# (Optional) แก้คำสะกดซ้ำซ้อนแบบง่าย ๆ — เพิ่มใน dict ตามต้องการ
replace_map = {
    "เรียบ เท่ (ดำ เทา น้ำเงิน)": "เรียบ เท่",
    "สีเรียบ เท่ (ดำ เทา น้ำเงิน)": "เรียบ เท่",
    "สีพาสเทล น่ารัก": "พาสเทล น่ารัก",
}
style_long["value"] = style_long["value"].replace(replace_map)

style_counts = (style_long["value"]
                .value_counts()
                .head(15)
                .rename_axis("Style / Color")
                .reset_index(name="Count"))
display(style_counts)
plot_bar(style_counts, y="Style / Color", title="Top Preferred Styles / Colors")

##------------------------------------------------------------
## 5) QUICK DRAFT INSIGHT
##------------------------------------------------------------
print("🔍  Draft Insight — Product Preferences")
print("• Bag-type อันดับ 1 :", top_bags.iloc[0]['Bag Type'])
print("• Size ฮิต :", size_counts.sort_values('Count', ascending=False).iloc[0]['Size (inches)'])
print("• Style/Color ฮิต :", style_counts.iloc[0]['Style / Color'])
# ===========================================================================
```

```python
# === STEP 5 : Price Sensitivity =========================================
import re, numpy as np
import matplotlib.pyplot as plt, seaborn as sns

# 5.1 ฟังก์ชันแปลงช่วงราคาจากข้อความ → ตัวเลข ------------------------
def price_range_to_tuple(txt):
    """
    '1,000 - 3,000' -> (1000, 3000, 2000)
    '2,500'         -> (2500, 2500, 2500)
    """
    if pd.isna(txt):
        return (np.nan, np.nan, np.nan)
    nums = list(map(int, re.findall(r"\d+", txt.replace(",", ""))))
    if len(nums) == 2:
        lo, hi = nums
        return (lo, hi, (lo + hi) / 2)
    elif len(nums) == 1:
        return (nums[0], nums[0], nums[0])
    else:
        return (np.nan, np.nan, np.nan)

# 5.2 สร้าง price_min / price_max / price_mid ---------------------------
df[["price_min", "price_max", "price_mid"]] = (
    df["preferred_price_range"]
      .apply(price_range_to_tuple)
      .apply(pd.Series)
)

# 5.3 Box-plot : price_mid × income_group -------------------------------
plt.figure(figsize=(7,4))
sns.boxplot(data=df, x="income_group", y="price_mid", showfliers=False)
plt.title("Preferred Mid-Price by Income Group")
plt.xlabel("Income Group (THB / month)")
plt.ylabel("Preferred Mid-Price (THB)")
plt.tight_layout()
plt.show()

# 5.4 ตารางค่าเฉลี่ยคะแนน price_* (Likert 1-5) ------------------------
price_cols = [c for c in df.columns if c.startswith("price_")]
price_means = df.groupby("income_group")[price_cols].mean().round(2)
display(price_means.style.background_gradient(cmap="YlGnBu", axis=None))

# 5.5 Draft Insight (safe version)
print("\n🔍  Draft Insight — Price Sensitivity")
for grp, row in price_means.iterrows():
    notnull = row.dropna()
    if notnull.empty:
        print(f"• {grp}: (ไม่มีข้อมูลคะแนน price_*)")
    else:
        top_col = notnull.idxmax()
        score   = notnull[top_col]
        print(f"• {grp}: ให้คะแนน '{top_col}' สูงสุด → {score}")

```

```python
# === STEP 8 : Brand Awareness (final) ===================================
import matplotlib.pyplot as plt, seaborn as sns
import numpy as np, pandas as pd

# 8.1 แยกกลุ่มรู้จัก / ไม่รู้จัก ----------------------------------------
def brand_know_status(txt):
    if pd.isna(txt):
        return np.nan
    txt = txt.strip().lower()
    if "ไม่รู้จัก" in txt:
        return "no"
    # ทุกข้อความที่เหลือให้เป็น yes  (รู้จัก / เคยเห็น / รู้จักดี)
    return "yes"

know_clean = df["know_tpartner"].apply(brand_know_status)

cnt_yes = (know_clean == "yes").sum()
cnt_no  = (know_clean == "no").sum()
pct_yes = cnt_yes / len(df) * 100
pct_no  = cnt_no  / len(df) * 100

print(f"✅ Respondents who KNOW the brand : {cnt_yes}  ({pct_yes:.1f} %)")
print(f"❌ Respondents who DON'T know    : {cnt_no}  ({pct_no:.1f} %)")

# 8.2 ช่องทางแรก (คิด % จากคนที่รู้จักเท่านั้น) -------------------------
first_chan = (
    df.loc[know_clean == "yes", "tpartner_first_channel"]
      .dropna()
      .apply(normalize_th)
)

chan_counts = (
    first_chan.value_counts()
      .rename_axis("First-touch channel")
      .reset_index(name="Count")
)

chan_counts["Percent"] = chan_counts["Count"] / cnt_yes * 100
display(chan_counts)

plt.figure(figsize=(7,4))
sns.barplot(y="First-touch channel", x="Percent", data=chan_counts)
plt.title("Where People First Heard About the Brand")
plt.xlabel("% of respondents who know the brand")
plt.ylabel("")
plt.tight_layout()
plt.show()

# 8.3 Draft Insight -------------------------------------------------------
print("\n🔍  Draft Insight — Brand Awareness")
if not chan_counts.empty:
    top_row = chan_counts.iloc[0]
    print(f"• ช่องทางที่คนรู้จักแบรนด์มากที่สุด : {top_row['First-touch channel']}  →  {top_row['Percent']:.1f}%")
else:
    print("• ยังไม่มีข้อมูลช่องทางแรก (tpartner_first_channel)")
print(f"• คนที่ไม่รู้จักแบรนด์เลย : {pct_no:.1f}% ของทั้งหมด")
```

```python
# === STEP 9 : Brand Image ==============================================
import re, unicodedata, collections
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns

# 9.1 รวมข้อความจาก 3 ฟิลด์ ---------------------------------------------
text_cols = ["first_impression", "tpartner_positioning", "tpartner_memorable_factor"]
df["text_all"] = df[text_cols].fillna("").agg(" ".join, axis=1)

# 9.2 ตัดคำไทย -----------------------------------------------------------
from pythainlp import word_tokenize
thai_stop = set([
    "ๆ","ค่ะ","คะ","ครับ","แล้ว","และ","คือ","ว่า","ของ","ให้","กับ",
    "ได้","ไป","ใน","เป็น","ก็","ที่","โดย","ๆ","มัน","ฉัน","เรา"
])

def clean_token(tok):
    tok = tok.strip().lower()
    tok = unicodedata.normalize("NFKC", tok)
    return re.sub(r"[^\u0E00-\u0E7Fa-zA-Z0-9]", "", tok)  # คงไทย+อังกฤษ+ตัวเลข

tokens = []
for txt in df["text_all"]:
    toks = [clean_token(t) for t in word_tokenize(txt, keep_whitespace=False)]
    tokens.extend([t for t in toks if t and t not in thai_stop and not t.isdigit()])

# 9.3 นับความถี่คำ --------------------------------------------------------
freq = collections.Counter(tokens)
top30 = pd.DataFrame(freq.most_common(30), columns=["word","count"])
display(top30)

plt.figure(figsize=(8,5))
sns.barplot(y="word", x="count", data=top30, palette="crest")
plt.title("Top 30 Keywords in Brand Image Responses")
plt.xlabel("Frequency")
plt.ylabel("")
plt.tight_layout()
plt.show()

# 9.4 แบ่งหมวดคำเชิงบวก / กลาง / ลบ ------------------------------------
positive_kw = {"วัยรุ่น","ทันสมัย","สวย","ทน","คุณภาพ","ใช้งานง่าย","คุ้มค่า"}
negative_kw = {"ไม่ทน","แพง","ไม่มั่นใจ","ซ่อมยาก","ขาด","หนัก"}
# เหลืออื่นๆ ถือเป็นกลาง
pos_cnt = sum(freq[w] for w in positive_kw if w in freq)
neg_cnt = sum(freq[w] for w in negative_kw if w in freq)
neu_cnt = sum(freq.values()) - pos_cnt - neg_cnt

sent_df = pd.DataFrame({
    "Tone":["Positive","Neutral","Negative"],
    "Count":[pos_cnt, neu_cnt, neg_cnt]
})
display(sent_df)

plt.figure(figsize=(4,4))
sns.barplot(data=sent_df, x="Tone", y="Count", palette=["#4caf50","#ffeb3b","#f44336"])
plt.title("Sentiment Tone of Brand Image")
plt.tight_layout()
plt.show()

# 9.5 Draft Insight -------------------------------------------------------
print("\n🔍  Draft Insight — Brand Image")
print(f"• คำบวกสูงสุด: {', '.join([w for w in positive_kw if w in freq][:3])}")
print(f"• คำลบสูงสุด : {', '.join([w for w in negative_kw if w in freq][:3])}")
print(f"  (Positive {pos_cnt} | Neutral {neu_cnt} | Negative {neg_cnt})")
```

```python
# === STEP 10 : Barrier to Purchase =====================================
import re, unicodedata, collections, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from pythainlp import word_tokenize

# 10.1 รวบรวมข้อความ -----------------------------------------------------
raw_barrier = df["reason_not_chosen"].dropna().tolist()
print(f"💬 มีคำตอบปลายเปิดทั้งหมด {len(raw_barrier)} รายการ")

# 10.2 ตัดคำ + ทำความสะอาด ---------------------------------------------
thai_stop = {"ๆ","ค่ะ","คะ","ครับ","ว่า","คือ","ของ","ให้","กับ","ได้","ไป","ใน","แต่","ยัง"}
def normalize(x):
    x = unicodedata.normalize("NFKC", x.lower())
    x = re.sub(r"[^\u0E00-\u0E7Fa-zA-Z0-9 ]", " ", x)  # คงไทย+eng+num
    x = re.sub(r"\s+", " ", x).strip()
    return x

tokens = []
for txt in raw_barrier:
    toks = [t for t in word_tokenize(normalize(txt), keep_whitespace=False)
            if t not in thai_stop and not t.isdigit()]
    tokens.extend(toks)

freq = collections.Counter(tokens)

# 10.3 กำหนด “Pain-Point Dictionary” ------------------------------------
pain_dict = {
    "ยังไม่ได้เที่ยว": ["ยังไม่ได้เที่ยว","ไม่ได้เดินทาง","ยังไม่เดินทาง"],
    "รีวิวน้อย/ไม่มั่นใจ": ["รีวิว","ไม่เห็นรีวิว","ไม่มั่นใจ"],
    "แพง/รอโปร": ["แพง","ราคา","รอโปร","รอลด","งบ"],
    "ไม่มีร้านให้ลอง": ["ไม่มีร้าน","ไม่มีหน้าร้าน","ไม่ได้ลองจับ"],
    "ไม่ชอบดีไซน์/สี": ["ดีไซน์","สี","ลาย"],
    "ขนาด/น้ำหนักไม่ตรง": ["ขนาด","ใหญ่","เล็ก","หนัก"],
}
pain_dict.update({
    "ยังไม่ได้เที่ยว": ["ยังไม่ได้เที่ยว", "ยังไม่มีโอกาส", "ไม่ได้เดินทาง"],
    "กระเป๋าเก่าทน": ["ใบเก่ายัง", "ใช้ใบเก่า", "เก่ายังไม่พัง"],
    "ชอบแบรนด์อื่น": ["ชอบแบรนด์อื่น", "สนใจแบรนด์อื่น", "มีแบรนด์ประจำ"],
    "คอนเทนต์ไม่ตรง": ["ไม่ชอบคอนเทนต์", "คอนเทนต์", "สื่อที่ปกติ", "ไม่ตรงกับลักษณะสื่อ"],
})

# 10.4 จัดหมวดหมู่ (แต่ละข้อความอาจเข้าหลายหมวด) ----------------------
cat_counts = collections.Counter()
for txt in raw_barrier:
    txt_norm = normalize(txt)
    matched = False
    for cat, kws in pain_dict.items():
        if any(kw in txt_norm for kw in kws):
            cat_counts[cat] += 1
            matched = True
    if not matched:
        cat_counts["อื่น ๆ"] += 1

bar_df = pd.DataFrame(cat_counts.items(), columns=["Barrier","Count"]).sort_values("Count", ascending=False)
display(bar_df)

# 10.5 กราฟ Treemap หรือ Bar – ใช้ barplot ง่าย ๆ ------------------------
plt.figure(figsize=(7,4))
sns.barplot(data=bar_df, x="Count", y="Barrier", palette="flare")
plt.title("Top Barriers to Purchase")
plt.xlabel("Number of Mentions")
plt.ylabel("")
plt.tight_layout()
plt.show()

# 10.6 Draft Insight ------------------------------------------------------
print("\n🔍  Draft Insight — Barrier to Purchase")
for idx, row in bar_df.head(5).iterrows():
    print(f"• {row['Barrier']}  →  {row['Count']} คน")
```

```python
# === STEP 11A : Persona จาก Demographic + Income ========================
import pandas as pd, textwrap

# 1. สรุปกลุ่มที่มีคนมากที่สุด
demo_tbl = (
    df.groupby(["gender", "income_group"])
      .size()
      .reset_index(name="Count")
      .sort_values("Count", ascending=False)
)
display(demo_tbl)

# 2. เลือก Top 3 กลุ่ม → สร้างการ์ด Persona เบื้องต้น
top3 = demo_tbl.head(3).to_dict("records")

print("\n🗂 Draft Persona (Demographic-Based)")
for i, p in enumerate(top3, 1):
    name = {
        ("หญิง","<15k"):  "Uni-Vivi",
        ("หญิง","15-30k"): "Starter-Sita",
        ("ชาย","<15k"):   "Budget-Beam",
    }.get((p["gender"], p["income_group"]), f"Persona {i}")
    
    print(f"\n{name}")
    print("-"*len(name))
    print(f"Gender   : {p['gender']}")
    print(f"Income   : {p['income_group']}")
    # เติม Insight ย่อจากขั้นก่อน ๆ
    key_need   = "Durability, Value-for-Quality" if p["income_group"]=="<15k" else "Design & Warranty"
    pain_point = "ยังไม่ได้เที่ยว / รีวิวไม่พอ" if p["income_group"]=="<15k" else "รอโปร / เปรียบเทียบแบรนด์"
    channel    = "TikTok → Shopee" if p["gender"]=="หญิง" else "Lazada Flash Sale"
    promo      = "ส่วนลดตรง ๆ" if p["income_group"]=="<15k" else "Bundle Gift"
    
    print(f"Top Needs: {key_need}")
    print(f"Pain-Pt  : {pain_point}")
    print(f"Channel  : {channel}")
    print(f"Promo    : {promo}")
```

```python
# === STEP 11B : Behaviour-Clustering ===================================
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# 1) เตรียมชุดฟีเจอร์ ---------------------------------------------------
feat_cols = [c for c in df.columns
             if c.startswith(("factor_", "price_", "channel_", "promo_"))]


X = df[feat_cols].fillna(df[feat_cols].median())   # กัน Missing
X_scaled = StandardScaler().fit_transform(X)

# 2) หา K ที่เหมาะ (Elbow + Silhouette) --------------------------------
sil = []; inert = []
K_range = range(2, 7)
for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init="auto")
    km.fit(X_scaled)
    inert.append(km.inertia_)
    sil.append(silhouette_score(X_scaled, km.labels_))

fig, ax = plt.subplots(1,2, figsize=(9,3))
ax[0].plot(K_range, inert, "-o");        ax[0].set_title("Elbow (Inertia)")
ax[1].plot(K_range, sil,  "-o");         ax[1].set_title("Silhouette")
for a in ax: a.set_xlabel("k"); a.set_xticks(K_range)
plt.tight_layout(); plt.show()

# === เลือก k=3 (ถ้า Elbow & Silhouette เห็นว่าดี) ======================
k = 3
kmodel = KMeans(n_clusters=k, random_state=42, n_init="auto")
df["cluster"] = kmodel.fit_predict(X_scaled)

print("📊 Cluster size\n", df["cluster"].value_counts())

# 3) สรุปคุณลักษณะแต่ละคลัสเตอร์ -------------------------------------
cluster_mean = (df.groupby("cluster")[feat_cols]
                .mean()
                .round(2))
display(cluster_mean.head())

# 4) จับชื่อ Persona ตาม pattern ของค่าเฉลี่ย --------------------------
cluster_name = {}
for c,row in cluster_mean.iterrows():
    if   row["promo_discount"] > 4.3:     cluster_name[c] = "Promo-Lover"
    elif row["factor_design"] > 4.4:      cluster_name[c] = "Design-Hunter"
    else:                                 cluster_name[c] = "Value-Seeker"

df["persona_b"] = df["cluster"].map(cluster_name)

# 5) สรุป Demographic + Cluster Label -----------------------------------
summary = (
    df.groupby(["persona_b","gender","income_group"])
      .size()
      .reset_index(name="Count")
      .sort_values("Count", ascending=False)
)
display(summary.head(10))

# 6) กราฟ Heat-map ค่าเฉลี่ย factor หลักในแต่ละ Persona ---------------
top_fact = ["factor_durability","factor_design","factor_weight",
            "price_value_for_quality","promo_discount"]
plt.figure(figsize=(7,4))
sns.heatmap(cluster_mean[top_fact]
            .rename(index=cluster_name),
            annot=True, cmap="YlGnBu", vmin=1, vmax=5)
plt.title("Key Drivers by Behaviour Persona")
plt.ylabel("Persona")
plt.tight_layout()
plt.show()

# 7) Draft Insight -------------------------------------------------------
print("\n🔍  Draft Insight — Behaviour Persona (k=3)")
for c,name in cluster_name.items():
    size = (df["cluster"]==c).sum()
    top_need = cluster_mean.loc[c].idxmax()
    print(f"• {name}  (n={size})  →  เจาะที่ '{top_need}' สูงสุด")
```
